```python
import json
```

```python
# Func to Load the data
def load_data(filename):
    with open (filename , "r") as f:
        data= json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
# Func to Clean & Struc the Data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five":5} #dict
    cleaned_data = [] #list
    unique_users = set()

    for user in data: #loop runned for all users given in data
        # Clean ratings-Data consistency 
        raw_rating = str(user["rating"]).strip().lower() #strip method-cancels spaces before or after|| #lower method-convert in lowercase
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating] 
        user["rating"] = raw_rating #corrected inconsistent data

        #Handle missing vals
        raw_age = user.get("age") #get method gets age if present , if not present then assigns None val to it
        if(raw_age == None):
            user["age"] = None #assigned None val for the missing age of users

        #Deduplication ----------------------------**************imp
        clean_name = user["name"].strip().lower() # Stripped extra spaces in user name & checked for lowercased "Alice" != "alice"
        if(clean_name in unique_users): # Check if the CLEAN name is in the set
            continue #break from the loop
            
        #if not
        unique_users.add(clean_name) #add clean name to the set
        cleaned_data.append(user) #assigned the cleaned user in cleaned data

    return cleaned_data
            
```

```python
clean_data(data)
```

```python
# Func to Get meaningful insights from data
data = clean_data(data)  # ensuring the cleaning function we fixed earlier is actually applied to the variable data before you run the insights.
def get_insights(data): # this is my cleaned data

    #avg rating 
    tot_rating = 0 #initializing total rating with 0
    for user in data:
        tot_rating += float(user["rating"]) #In each loop iteration, the user's rating is retrieved, converted into a decimal number (float()), and then added to the running total stored in the tot_ratings variable.

    print(f"avg rating = {tot_rating/len(data)}")
    
    # %percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) <3):
            poor_rating +=1 #When the condition is met, it increments the poor_ratings counter by exactly 1.
                            #if cond is false then poor_rating does not update, stays the same.
    
    print(f" % of users with poor rating = {poor_rating/len(data) * 100} % ")
            
```

```python
get_insights(data)
```

```python
# Func to build recommendation feature
data = clean_data(data) 
def get_recommendation(data):
    recommendation = [] # dict stored inside list [ {}, {}, {}] , creates empty container (a list) so it's ready to be filled.

    for user in data:
        curr_recomm = {} 
        curr_recomm["name"] = user["name"] #Recommending Product name to users according to their rating below

        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"

        recommendation.append(curr_recomm) #.append(curr_recomm) takes that finished piece of work and safely adds it to mylist.
    return recommendation 
```

```python
 get_recommendation(data)
```
