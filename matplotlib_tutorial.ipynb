```python
# Comparing 2 sets of data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]


import numpy as np 
x = np.arange(len(years)) # shows index of years 
#we customize width. Width controls how thick each bar is. Usually ranges from 0 to 1. [Default value = 0.8]
width = 0.4


plt.bar(x - width/2 , oscar_revenue,  width, label = "Oscar Movies" ) # x - 0.2 (Shifting left)--> logic written in notes refer
plt.bar(x + width/2, non_oscar_revenue,  width,  label = "Non-Oscar Movies") # x + 0.2 (Shifting right)

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()

plt.xticks(x, years) # to show x-axis labels
```

```python
#SCATTER PLOTS 
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# plot
# plt.scatter(age, blood_pressure, color="red")

plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

colors = ["green" if x < 135 else "red" for x in bp]
sizes = bp # more bp , more big size of the circle
# plt.scatter(age, blood_pressure, color = colors, s= sizes, alpha= 0.5) #alpha = 1 (by default) (darkest)----->here 0.5 is translucent

# plt.scatter(age, blood_pressure, c=blood_pressure, cmap="plasma")
plt.scatter(age, bp, c=bp, marker = "^", s= sizes, cmap="OrRd")# makes more sense ---> cmap = colormap check docs
# plt.scatter(age, blood_pressure, c=[5]*len(age), cmap="plasma")

plt.colorbar(label="BP")  # gives colorbar at side


#plt.annotate("text", xy = (x_point, y_point), xytext = (x_text, y_text))  ---> writes person a , b , c ,d at respective places
for i in range(len(people)):
    plt.annotate(people[i], xy = (age[i],bp[i]), xytext = (age[i] +1 ,bp[i] + 0.5) ) #xytext - > where we want our text on graph

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

# plt.pie(amounts, labels=expenses) ---> plots the pie

# set style
plt.style.available
plt.style.use('dark_background')
plt.style.use('default')

#set colors
# colors = ["purple", "green", "blue", "pink", "yellow"]
# plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', colors=colors)

explode = [0 , 0, 0, 0.2, 0]

# wedgeprops & customization 
plt.pie(amounts, labels=expenses, autopct = "%1.1f%%", # autopct = "%1.1f%%"----> shows %-age of each slice
           wedgeprops = {
               "edgecolor" : "black",     #border color
               "linewidth" : 1,           #border thickness
               "linestyle" : "--",        #border dotted line or any style refer docs
               # "fill" : False           # ' unfills ' color ---> removes color
           },

        explode = explode,                #explode --> highlights the desired slice by pulling it out
        startangle = 360,                 # rotates chart by angle (90 , 180 , 360 , etc)
        shadow = True                     # gives 3D effect 
       )
```

```python
# Histograms--> numerical data distribution

# generate random data
import numpy as np
np.random.seed(0)

scores = np.random.normal(loc=70, scale=10 ,size=100) #normal distribution ( 100 random vals, 70 mean val, 10 standard deviation)
#bell-curve
plt.hist(scores, color = "skyblue", edgecolor = "black", bins = 20) #bins->ranges -> cal freq of each bin
plt.title("Distribution of exam scores")
plt.xlabel("Marks")
plt.ylabel("Number of students")
#print(scores)
```

```python
# custom bins
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

# custom bins
bins = [0, 30, 70, 90, 100]
plt.hist(marks, color = "skyblue", edgecolor = "black", bins = bins)

plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)

# particular boundary---> 70-90 range --> left range 70-inclusive --> right range 90-exclusive ( except for 100 at last [90-100])
#bins->continuos range (max-min ) 
#eg:- range = (0-100) --> max-min = 100-0 = 100 ---> if bins = 10, then 0 to 100 gets divided in 10 equal val, 0-10,10-20,20-30,...90-100
```

```python
# axv line 
#->draws a vertical ref line on a plot to mark cutoffs, averages, pass-marks or thresholds.
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

# custom bins
bins = [0, 30, 70, 90, 100]
plt.hist(marks, color = "skyblue", edgecolor = "black", bins = bins)

# line at passing marks - 33
plt.axvline(
    x = 33,
    color = "red",
    linestyle = "--",
    linewidth = 2,
     label="Pass Mark (33)"
)

plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)
```

```python
# Multiple datasets Histogram
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

plt.hist(fraud_transactions, color = "red", alpha = 0.5, edgecolor = "black", bins = 20, label = "Fraud")
plt.hist(legit_transactions, color = "green", alpha = 0.5, edgecolor = "black", bins = 20, label = "Legit")

plt.legend()
plt.grid(True)

plt.xlabel('Transaction Amount ($)')
plt.ylabel("Frequency")
plt.title('Transaction Amount Distribution: Legitimate vs Fraudulent')

#legit transactions are left skewed -small transactions
#we can see fraud transactions are mostly on whole and larger amounts 

#avoid mistakes ---> too few bin vals - lacks structure, ---> too many bin vals - creates noise --> hence balance bins
#don't use histogram for categorical data (eg: expenses in company)----> use hist for numerical data
```
